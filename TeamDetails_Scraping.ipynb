```python
import requests
import csv
from bs4 import BeautifulSoup
from selenium  import webdriver
import pandas as pd
import numpy as np
driver = webdriver.Chrome(executable_path='chromedriver.exe')
driver.get('https://www.prokabaddi.com/stats/49-96-total-points-scored-statistics')
element = driver.find_element_by_id('load_more')
element.click()
soup = BeautifulSoup(driver.page_source)
for item in soup.find_all("div",class_="sipk-lb-playerName"):
    [elem.extract() for elem in soup.find("div")]
    f = open("parsed.txt", "a")
    f.write(item.text)
    f.close()
teamDetails = pd.read_csv('parsed.txt',encoding='ISO-8859-1',sep='\t',names=['TeamName'])
print(teamDetails)
    


import os
if os.path.exists("parsed.txt"):
    os.remove("parsed.txt")
    
   

   


```
