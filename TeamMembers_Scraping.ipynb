```python
import requests
import csv
from bs4 import BeautifulSoup
from selenium  import webdriver
import pandas as pd
import numpy as np
driver = webdriver.Chrome(executable_path='chromedriver.exe')
driver.get('https://www.prokabaddi.com/teams/bengal-warriors-profile-4')
soup = BeautifulSoup(driver.page_source)
for item in soup.find_all("div",class_=["si-txt1"]):
    [elem.extract() for elem in soup.find("span")]
    f = open("parsed.txt", "a")
    f.write("\n"+item.text+"\n")
    f.close()
firstNames = pd.read_csv('parsed.txt',sep="\n",names=['firstNames'])

import os
if os.path.exists("parsed.txt"):
    os.remove("parsed.txt")
    
    
   

   


```

```python
soup = BeautifulSoup(driver.page_source)
for item in soup.find_all("div",class_=["si-txt2"]):
    [elem.extract() for elem in soup.find("span")]
    f = open("parsed.txt", "a")
    f.write("\n"+item.text+"\n")
    f.close()
lastNames = pd.read_csv('parsed.txt',sep="\n",names=['lastNames'])

import os
if os.path.exists("parsed.txt"):
    os.remove("parsed.txt")

   
```

```python
dfObj=pd.DataFrame(data=[])
dfObj['PlayerName'] = firstNames[firstNames.columns[0]]+"  "+lastNames[lastNames.columns[0]]
dfObj
```

```python
dfObj['TeamName'] = 'Bengal Warriors'
for i in range(0, len(dfObj)):
    dfObj[dfObj.columns[1]]= 'Bengal Warriors'
    
dfObj
```
